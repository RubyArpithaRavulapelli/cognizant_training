```python
import pandas as pd
dataJoin=pd.DataFrame({
    "ID":[1,2,3],
    "NAME":["Arpitha","Ruby","Sweety"]})
print(dataJoin)
```

```python
dataJoin2=pd.DataFrame({
    "ID":[1,2,3],
    "RANK":[1,2,3]})
print(dataJoin2)
```

```python
leftJoin=pd.merge(dataJoin,dataJoin2,on="ID",how="left")
print(leftJoin)
```

```python
rightJoin=pd.merge(dataJoin,dataJoin2,on="ID",how="right")
print(rightJoin)
```
