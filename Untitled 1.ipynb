```python
import random
developers=list(range(1,43))
testers=list(range(43,62))
managers=list(range(62,100))
sample_dev=random.sample(developers,8)
sample_test=random.sample(testers,2)
sample_marg=random.sample(managers,4)
final_sample=sample_dev+sample_test+sample_marg
print(final_sample)
```

```python
clusters={
    "Godavarikhani":[1,2,3],
    "Ramagundam":[4,7,8],
    "Vizag":[11,12,32]
}
select_clusters=["Vizag","Godavarikhani"]
sample=[]
for cluster in select_clusters:
    sample.extend(clusters[cluster])
    print("cluster sample:",sample)
```

```python
employee={
    "A":12,
    "B":23,
    "C":45,
    "D":56
}
sample=[name for name,exp in employee.items() if exp>=7]
print("Judgment sample:",sample)
```

```python
networks={
    "A":["B","C"],
    "B":["E"],
    "C":["K"]
}
sample=["A"]
sample.extend(networks["A"])
sample.extend(networks["B"])
print("snowball sampling:",sample)
```
